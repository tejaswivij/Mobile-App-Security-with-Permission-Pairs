```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import BernoulliNB
import lightgbm as lgb
from xgboost import XGBClassifier
```

```python
def get_models():
    models = list()
    models.append(LogisticRegression())
    models.append(SGDClassifier())
    models.append(KNeighborsClassifier())
    models.append(DecisionTreeClassifier())
    models.append(ExtraTreeClassifier())
    models.append(GaussianNB())
    models.append(AdaBoostClassifier())
    models.append(RandomForestClassifier())
    models.append(GradientBoostingClassifier())
    models.append(MLPClassifier())
    models.append(BernoulliNB())
    models.append(lgb.LGBMClassifier())
    models.append(XGBClassifier())
    return models
```

```python
models = get_models()
len(models)
```

```python
for model in models:
  print(model,"function avaliable")
```

```python
from sklearn.metrics import *
from timeit import default_timer as timer
```

```python
# k fold and stratified k fold
from sklearn.model_selection import *
from timeit import default_timer as timer
from sklearn import metrics
import statistics
skf = StratifiedKFold(n_splits=10)
```

```python
for model in models:
    roc_scores=[]
    accuracys = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    results = []
    predicted_y = []
    expected_y = []
    cv_score_mean=0
    t = timer()
    scoring ='accuracy'
    start = timer()
    count = 0
    for train_index, test_index in skf.split(X1, y):
        count+=1
        print("Model : ", model)
        x_train, x_test = X1.loc[train_index], X1.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        #x_train = sc.fit_transform(x_train)
        #x_test = sc.fit_transform(x_test)
        package = model.fit(x_train, y_train)
        y_predict = package.predict(x_test)
        cm = confusion_matrix(y_test, y_predict)
        print("Fold : ",count)
        tn,fp,fn,tp = cm.ravel()
        print("True_negative : ",tn)
        print("false_positive : ",fp)
        print("false_negative : ",fn)
        print("True_positive : ",tp)
        precision = tp / (tp + fp)
        #print("Precision : ",precision)
        precision_scores.append(precision)
        recall = tp / (tp + fn)
        #print("Recall : ",recall)
        recall_scores.append(recall)
        specificity = tn / (tn+fp)
        print("Specificity : ",specificity)
        accuracy = accuracy_score(y_test, y_predict)
        print("Accuracy : ",accuracy)
        accuracys.append(accuracy)
        #print("F1_score : ",f1_score(y_test,y_predict))
        f1=f1_score(y_test,y_predict)
        f1_scores.append(f1)
        roc_score=roc_auc_score(y_test,y_predict)
        #print("Roc_score : ",roc_score)
        roc_scores.append(roc_score)

        print("\n")
    print("Model : ", model)
    print("folds : ", count)
    tns,fps,fns,tps = 0,0,0,0

    end = timer()
    print("time elapsed : ",end - start)
    print("Accuracy_score_Max : ",max(accuracys))
    print("Accuracy_score_mean : ",statistics.mean(accuracys))
    print("ROC_score_max : ",max(roc_scores))
    print("ROC_score_mean : ",statistics.mean(roc_scores))
#    print("CV_results_max : ",cv_results.max())
#    print("CV_results_mean : ",cv_results.mean())
#    print("std_deviation : ",cv_results.std())
    avg_accuracy = np.mean(accuracys, axis=0)
    print("avg_accuracy : ",avg_accuracy)
    std_accuracy = np.std(accuracys, axis=0)
    print("std_accuracy : ",std_accuracy)

    avg_precision = np.mean(precision_scores, axis=0)
    print("avg_precision : ",avg_precision)
    std_precision = np.std(precision_scores, axis=0)
    print("std_precision : ",std_precision)

    avg_recall = np.mean(recall_scores, axis=0)
    print("avg_recall : ",avg_recall)
    std_recall = np.std(recall_scores, axis=0)
    print("std_recall : ",std_recall)

    avg_f1 = np.mean(f1_scores, axis=0)
    print("avg_f1: ",avg_f1)
    std_f1 = np.std(f1_scores, axis=0)
    print("std_f1 : ",std_f1)

    print("\n")
    #print(f'\nModel: {model}\naccuracy_score_max: {max(accuracys)}\naccuracy_score_mean: {statistics.mean(accuracys)}\nroc_scores_max: {max(roc_scores)}\nroc_scores_mean: {statistics.mean(roc_scores)}\nlist of scores: {cv_results}\ncv_results_max: = {cv_results.max()}\ncv_results_Mean: {cv_results.mean()}\nStd Dev: {cv_results.std()}\n\n')
    
```
