```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Generated values (mu Â± k*sigma)
generated_values = np.array([19.25126958, 13.09556564, 6.9398617, 0.78415776, -5.37154618, -11.52725012]).reshape(-1,1)

# Normalize values to [0.1, 1] to avoid zero
scaler = MinMaxScaler(feature_range=(0.1, 1.0))
normalized_values = scaler.fit_transform(generated_values).flatten()

# Sort folds by importance
fold_priority = [5, 4, 3, 2, 1, 0]
fold_features_list = [
    features_by_fold_count.get(5, []),
    features_by_fold_count.get(4, []),
    features_by_fold_count.get(3, []),
    features_by_fold_count.get(2, []),
    features_by_fold_count.get(1, []),
    list(never_malware_features)  # 0-fold features
]

# Assign normalized values to features based on fold priority
S_map = {}
for i, features in enumerate(fold_features_list):
    if features:
        # Use normalized value corresponding to fold priority
        value = normalized_values[i]
        for feat in features:
            S_map[feat] = value

print("\nSignificance (normalized) values for all features:")
for feat, val in S_map.items():
    print(f"{feat}: {val:.4f}")
```

```python
#compute the risk score
# Compute Pr^m and Pr^n for each permission pair
Pr_m = edu_sorted[edu_sorted['class']==1].iloc[:, :-1].mean()  # probability in malware
Pr_n = edu_sorted[edu_sorted['class']==0].iloc[:, :-1].mean()  # probability in benign

risk_scores = {}

for feat in S_map:
    S = S_map[feat]                   # significance
    p_m = Pr_m.get(feat, 0)           # malware probability
    p_n = Pr_n.get(feat, 0)           # benign probability

    # Weight function
    W = (p_m ** 2) / (p_m ** 2 + p_n ** 2 + 1e-8)  # avoid division by zero

    # Risk score formula
    term = (1 + np.sqrt((1 - p_n) * p_m) - np.sqrt((1 - p_m) * p_n)) / 2
    R = S * W * term
    risk_scores[feat] = R

# Convert to DataFrame for easy viewing
risk_df = pd.DataFrame.from_dict(risk_scores, orient='index', columns=['RiskScore']).sort_values(by='RiskScore', ascending=False)

print(risk_df)
```
